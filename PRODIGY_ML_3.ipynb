```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
df=pd.read_csv('data.csv')
df.head()
```

```python
df.country.values
```

```python
df1=df.drop(['date','sqft_lot','waterfront','statezip','street','country'],axis='columns')
df1.head()
```

```python
df1.floors=df1.floors.round()
df1.bathrooms=df1.bathrooms.round()
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df1.price.describe()
```

```python
df1 = df1.drop(df1[df1['price'] == 0].index, axis=0)
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1.city=le.fit_transform(df1.city)
df1.head()
```

```python
xtr,xt,ytr,yt=train_test_split(df1.drop(['price'],axis='columns'),df1.price,test_size=0.2)
xtr.shape
```

```python
xt.shape
```

```python
model=LinearRegression()
model.fit(xtr,ytr)
model
```

```python
model.predict(xt)
```

```python
model.score(xt,yt)
```
